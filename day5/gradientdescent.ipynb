```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------
# DATA
# ---------------------
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([5, 6, 10, 13, 11], dtype=float)

m = len(X)

# Initial parameters
b0 = 3
b1 = 1.9

# Learning rate
alpha = 0.01

# Store for plotting
errors = []

# ---------------------
# Mean Squared Error
# ---------------------
def mse(b0, b1):
    preds = b0 + b1 * X
    return np.mean((preds - y)**2)

# ---------------------
# Gradient Descent for multiple iterations
# ---------------------
num_iters = 100

for i in range(num_iters):
    preds = b0 + b1 * X
    
    # Gradients
    db0 = (1/m) * np.sum(preds - y)
    db1 = (1/m) * np.sum((preds - y) * X)
    
    # Update
    b0 = b0 - alpha * db0
    b1 = b1 - alpha * db1
    
    # Track error
    errors.append(mse(b0, b1))

print("Updated b0:", b0)
print("Updated b1:", b1)

# ---------------------
# Plot Error Curve
# ---------------------
plt.figure(figsize=(8,4))
plt.plot(errors, marker='o')
plt.xlabel("Iteration")
plt.ylabel("MSE")
plt.title("Error Curve (Loss vs Iterations)")
plt.grid(True)
plt.show()
```
